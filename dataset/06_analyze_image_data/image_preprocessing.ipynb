```python
import cv2
import os
import glob
import shutil
from tqdm.notebook import tqdm
```

```python
#라벨목록
folder_list = os.listdir('D:/marytalk/finish')
```

```python
folder_list
```

```python
#이미지들 경로, 이름 정리
for label in folder_list :
    files = glob.glob(f'D:/marytalk/finish/{label}/*/*')
    imagefiles = [f for f in files if f.endswith(('.png','.jpg'))]
    for i, file in enumerate(imagefiles) :
        label2 = label[:]
        dir_ = f'images/{label}/'
        if not os.path.isdir(dir_):
            os.makedirs(dir_)
        filename = dir_+label2+'_'+str(i).rjust(4,'0')+'.png'
        shutil.copy(file.replace('\\','/'), filename)
```

```python
#카테고리별 이미지 수 확인
rootdir='images/'
for f in folder_list:
    path = rootdir+f
    print(f+' :',len(next(os.walk(path))[2]))
```

```python
#이미지 폴더 내의 모든 png, jpg의 filepath를 리스트에 저장
def imagepath(path) :
    folderlist = glob.glob(path)
    imagefiles = [f for f in folderlist if f.endswith(('.png','.jpg'))]
    return imagefiles

#리스트에 저장된 이미지의 height, width를 튜플로 묶어서 리스트에 저장
def imageshape(path : str) :
    h_w = []
    imagefiles = imagepath(path)
    for f in tqdm(imagefiles) :
        img = cv2.imread(f)
        
        if img is None :
            os.remove(f)
        else :
            h = img.shape[0]
            w = img.shape[1]
            h_w.append((h, w))
            
    return h_w
```

```python
h_w = imageshape('images/*/*')
```

```python
#{(height, width):count} 형태로 사이즈 별 개수를 저장
shape_cnt = {}
for s in h_w :
    if s not in shape_cnt.keys():
        shape_cnt[s] = 1
    else :
        shape_cnt[s] += 1
```

```python
#x[1] = value
#값을 기준으로 내림차순하여 튜플 형태로 추출
sorted(shape_cnt.items(), reverse=True,  key=lambda x : x[1])
```

```python
#이미지 height = 225, width = 225로 지정
```

```python
def imageResize(folder_list: list):
    rootdir='original/'
    for c in folder_list :
        path = rootdir+c
        files = next(os.walk(path))[2]
        
        for f in files :
            resized_path = f'images/{c[3:]}/'
            if not os.path.isdir(resized_path):
                os.makedirs(resized_path)
            
            img = cv2.imread(path+'/'+f)
            shape_ = img.shape
            if shape_[0]<225 or shape_[1]<225 :
                resized = cv2.resize(img, dsize=(225, 225)
                                     , interpolation=cv2.INTER_CUBIC)
            else :
                resized = cv2.resize(img, dsize=(225, 225)
                                     , interpolation=cv2.INTER_AREA)
            
            cv2.imwrite(resized_path+f,resized)
```

```python
imageResize(folder_list)
```
