````python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt
import cv2
import os
from tqdm.notebook import tqdm
````

모델 구축시 중요한 것은 학습시킬 데이터입니다. <br>
우리는 헤어데이터를 60만장가량 크롤링했었으나 쓰레기 데이터의 양이 많은 문제로 얼굴부분만 다시 크롤링한 후 수작업으로 이미지를 걸러내어 데이터의 양이 심각하게 부족하게되어 어려움에 처하게됩었습니다.<br> 이를 위해서 이미지를 증식하는 모델을 구축합니다.

ImageDataGenerator는 케라스에서 제공하는 이미지 증식을 위한 라이브러리 입니다.

````python
datagen = ImageDataGenerator(
    rotation_range = 20
    ,width_shift_range = 0.2
    ,height_shift_range =0.2
    ,fill_mode='nearest'
)
````

````python
# ImageDataGenerator?
````

````python
categories = ['bobcut','babyperm','wolfcut','shagicut','spinswallow','chuffybang']
````

````python
for i in tqdm(categories):
    path = './hair_categories/'+i+'/'
    image_list = os.listdir(path)
    
    # 저장할 경로 입력
    save_dir = './hair_categories/'+i+'/'
    
    # 증식할 이미지 개수 지정
    augment_number = 110
    for f in image_list:
        image_filename = path+f
        image = np.expand_dims(plt.imread(image_filename), 0)
        datagen.fit(image)
        for x, val in zip(datagen.flow(
            image     # input image
            ,save_to_dir = save_dir        # augmented images will be saved here
            ,save_prefix=i                 # the augmented image name will have prefix 'aug'
            ,save_format='png'),range(augment_number)) :
            pass   
````
