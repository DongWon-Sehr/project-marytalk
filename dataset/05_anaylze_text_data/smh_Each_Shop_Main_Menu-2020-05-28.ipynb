```python
from tqdm.notebook import tqdm
```

```python
# 스타일 컬럼 생성
each_shop_review['style'] = np.NaN
```

```python
for i in tqdm(range(len(each_shop_review))):
    # 데이터프레임의 i번째 리뷰를 tmp 변수에 담는다.
    tmp = each_shop_review['review'].iloc[i]
    
    # tmp에서 keyword를 찾아서 result변수에 담는다.
    result = re.findall(keyword,tmp)

    # count 빈 리스트 생성
    count = []
    # set을통해 중복제거한 result를 key로 하나씩 뺀 후
    for key in set(result):
        # 각각의 키워드와 키워드의 갯수를 튜플로 count리스트에 넣어준다.
        count.append((key, result.count(key)))
        
    # 키워드의 갯수가 reverse를 통해서 내림차순 (큰 순서대로) 정렬해준다.
    sorted_count = sorted(count, key=lambda x:x[1],reverse=True)
    print(sorted_count)
    
    # 빈 리스트생성
    tmp_list=[]
    # 정렬된 리스트에서 각각의 튜플을 꺼낸다.
    for sorted_tuple in sorted_count:
        # 각각의 튜플의 키워드부분을 추출해 tmp_list에 넣어준다.
        tmp_list.append(sorted_tuple[0])
    # 추출한 키워드들을 콤마로 조인해서 문자열변환해준다.    
    str_count =','.join(tmp_list)    
    
    # 데이터프레임의 메뉴컬럼 해당 로우에 키워드문자열을 넣어준다.
    each_shop_review['style'].iloc[i] = str_count
print("***************************Finish**************************************")
```
