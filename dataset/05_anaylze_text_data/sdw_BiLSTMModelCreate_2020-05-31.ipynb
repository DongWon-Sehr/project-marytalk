```python
# TPU gRPC 접근 URI
import os
TPU_PATH = f"grpc://{os.environ['COLAB_TPU_ADDR']}"

resolver = tf.distribute.cluster_resolver.TPUClusterResolver(tpu=TPU_PATH)
tf.config.experimental_connect_to_cluster(resolver)
tf.tpu.experimental.initialize_tpu_system(resolver)
```

```python
strategy = tf.distribute.experimental.TPUStrategy(resolver)
```

```python
%%time

from anago.wrapper import Sequence
import datetime

print(f"\n[START TIME] {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S.%f')[:-2]}\n")
model = Sequence()

with strategy.scope():
    model.fit(x_train=tokens
            ,y_train=tags
            ,epochs=4
            ,batch_size=32
            ,shuffle=True)

print(f"\n[END TIME] {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S.%f')[:-2]}\n")
```
