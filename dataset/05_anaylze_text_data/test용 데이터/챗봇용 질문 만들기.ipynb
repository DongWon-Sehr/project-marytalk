```python
korean_re = re.compile('[^ ㄱ-ㅣ가-힣]+')
```

```python
def extract_last_word(text):
    text = korean_re.sub('',text)
    text = text.split()[-1]
    return text
```

```python
last_words = question_tag.sentence.apply(extract_last_word).tolist()
print(last_words)
```
