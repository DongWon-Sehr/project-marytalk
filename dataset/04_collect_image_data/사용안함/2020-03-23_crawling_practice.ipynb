```python
from bs4 import BeautifulSoup
import requests
```

```python
html_t = "https://search.naver.com/search.naver?where=post&sm=tab_jum&query=주노헤어+강남점"
# crawling할 URL을 html_t 변수에 저장해둔다.
response = requests.get(html_t)
# requests 모듈을 통해 html_t의 소스를 전부 긁어오고 response 변수에 저장한다.
soup = BeautifulSoup(response.text, 'html.parser')
# BeautifulSoup을 통해 response의 text를 soup변수에 저장하고 그 방식은 html.parser로 한다.
                                #사실 왜 html.parser를 쓰는지는 아직 파악하지 못했다.
soup
# 긁어온 것을 찍어보자
# ===================================================================================================================================
# 아래는 html_t의 selector를 복사하여 붙여넣은 것이다.
# sp_blog_1 > dl > dt > a
```

```python
type(soup)
#soup는 BeautifulSoup이라는 객체로 메모리에 생성(load)되었다.
```

```python
salon = soup.select('#main_pack > div.blog.section._blogBase._prs_blg > ul > li > dl > dt')
# soup.select에서 soup의 selector인 #main_pack > div.blog.section._blogBase._prs_blg > ul > li > dl > dt를 가져온다.
salon
# selector를 가져온 salon을 찍어보자.
```

```python
blist = []
# 검색결과의 URL을 담기 위해 빈 리스트를 생성했다.
for hairs in salon: # salon의 요소들을 하나씩 hairs라는 변수에 대입해준다.
    if not hairs.a == None: # 만일 hairs의 a태그가 None이 아니라면 실행되는 코드
        mylink = hairs.a # salon의 각 요소를 담은 hairs 변수의 a 태그들을 mylink라는 변수에 저장해두었다.
        print(mylink.attrs['href']) # attrs 메소드를 통해 salon의 key가 href인 value를 가져온다.
        # 아시다시피 href는 실질적인 링크를 가지고있기 때문에 URL을 받을 수 있게 된다.
        blist.append(mylink.attrs['href'])
        # 쓸모는 없지만 그래도 혹시 모르니 생성해두었던 빈 리스트에 저장하고 개수를 세어보자.
```

```python
len(blist)
# 놀랍게도 한 페이지의 검색 결과인 10개의 URL이 저장되어 있었다!
```

```python
# 웹 서치/크롤링을 자동화하기 위해서 selenium을 사용한다고 합니다.
# 저 역시도 구글링한 것을 바탕으로 간단한 자동화를 진행해보았습니다.
from selenium import webdriver
import time
```

# Selenium을 통해 웹 크롤링 자동화하기
---

1. Selenium 자동 탐색을 위해서는 크롬 드라이버(다른 드라이버도 가능)가 필요합니다. (크롬과 크롬 드라이버는 다른 파일입니다.)
2. Chrome Driver는 구글에 검색하면 다운로드 할 수 있습니다.
3. 크롬 드라이버 다운로드 창에서 win32용 드라이버를 다운로드하고, 압축을 풀어줍니다.
4. 압축을 풀 경로에 유의해주시고, 압축 해제가 완료되면 ChromeDriver가 있는 경로를 복사해주세요.
5. 그리고 driver라는 변수 선언 후 webdriver.Chrome('복사한 경로')를 대입해주세요.
6. 경로의 역슬래쉬(\\)를 슬래쉬(/)로 바꿔주세요. 역슬래쉬는 윈도우에서만 사용하는 경로 문자이기 때문에 역슬래쉬를 그대로 사용하면 인식을 못하게 됩니다.

```python
# driver = webdriver.Chrome('C:/Users/82106/Downloads/chromedriver_win32/chromedriver.exe')
# chromedriver의 path를 줄 때는 역슬래쉬가 아닌 그냥 슬래쉬(/)를 주어야한다!
# driver.get('https://www.naver.com/')
# 변수 driver를 통해 네이버 홈페이지로 이동하는 코드입니다.
```

```python
driver = webdriver.Chrome('C:/Users/82106/Downloads/chromedriver_win32/chromedriver.exe')
# chromedriver의 path를 줄 때는 역슬래쉬가 아닌 그냥 슬래쉬(/)를 주어야한다!
driver.get('https://www.naver.com/') # driver.get(URL) 드라이버를 킨다.
time.sleep(6.1) # Let the user actually see something after 6.1 seconds!
search_box = driver.find_element_by_name('query') # element name이 query인 것을 찾아 search_box에 대입한다.
search_box.send_keys('주노헤어 강남점') # send_keys(검색어 지정)로 search_box에 검색어를 지정(할당)한다.
search_box.submit() # submit이 실행되면 실질적인 검색을 한다.
#time.sleep(4.5) 
# quit()이 있다면 종료 전 4.5초를 쉬라는 의미인데 계속 켜둘꺼면 검색 후 sleep을 걸어두는 것은 의미가 없다.
# driver.quit()
# 드라이버를 종료하는 코드 (자동으로 꺼진다)
```
