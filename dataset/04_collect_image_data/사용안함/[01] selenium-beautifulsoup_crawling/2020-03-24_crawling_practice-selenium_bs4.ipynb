```python
driver = webdriver.Chrome('C:/Users/82106/Downloads/chromedriver_win32/chromedriver.exe')
url = 'https://www.naver.com/'
driver.get(url)
```

```python
time.sleep(5.384173)
search_box = driver.find_element_by_name('query')
search_box.send_keys('여자 헤어')
search_box.submit() 
```

```python
blog_tab = driver.find_element_by_css_selector('#lnb > div > div.lnb_menu > ul > li.lnb3 > a')
# 네이버 HTML소스로 블로그 탭의 xpath를 copy한 후 대입하였습니다.
time.sleep(7.298)
# 우리는 봇이 아니기 때문에 n초간 잠시 쉬는 시간을 가지는 겁니다.
blog_tab.click()
# click이라는 메소드가 blog_tab에 저장된 xpath를 찾아 클릭을 해줍니다.
```

```python
req = driver.page_source
req
```

```python
soup = BeautifulSoup(req, 'html.parser')
type(soup)
```

```python
salon = soup.select('#main_pack > div.blog.section._blogBase._prs_blg > ul > li > dl > dt')
salon
```

```python
zlist = []
for hshop in salon:
    if not hshop.a == None:
        blink = hshop.a
        print(blink.attrs['href'])
        zlist.append(blink.attrs['href'])
```
